```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/homeprices.csv")
df
```

Plot the distribution


```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
reg=linear_model.LinearRegression()
#fit means like training the model with available data
reg.fit(df[['area']],df['price'])
```

#Predict the home price

```python
reg.predict([[3300]])
```

```python
#coefficient of variables
```

```python
reg.intercept_
```

```python
# y=m*x+b
135.78767123*3300+180616.43835616432
```



---



```python
reg.predict([[5000]])
```

```python
135.78767123*5000+180616.43835616432
```



---



```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

# Save the model

```python
#write mode
# reg is your trained model (e.g., LinearRegression, RandomForest, etc.)
# pickle.dump(reg, f) serializes the model and saves it into a file named model_pickle.
import pickle
with open('model_pickle','wb') as f:
  pickle.dump(reg,f)
```

```python
#read mode 'rb'  read binary mode
# 'model_pickle'  the file where you earlier stored your model
import pickle
with open('model_pickle','rb') as f:
  mp=pickle.load(f)
```

```python
mp.predict([[5000]])
```
