Build a model to predict house prices based on size (sqft). Data provided in dictionary format. Predict price for 2500 sqft house.

Predict salary based on years of experience. Data: Experience [1, 2, 3, 5, 7, 10], Salary [30000, 35000, 45000, 55000, 65000, 80000]. Calculate the equation and predict salary for 6 years experience.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model

df = pd.DataFrame({
    'experience': [1, 2, 3, 5, 7, 10],
    'salary': [30000, 35000, 45000, 55000, 65000, 80000]
})
reg_3=linear_model.LinearRegression()
reg_3.fit(df[['experience']],df['salary'])
prediction=reg_3.predict([[6]])
m=reg_3.coef_[0]
intercept=reg_3.intercept_
print(f"Equation: {m:.2f}* experience + {intercept:.2f}")
print(f'Predicted salary for 6 years {prediction[0]:.2f}')
```

```python
reg_3.coef_
```

```python
reg_3.intercept_
```

```python
5552.3255814*6+25755.813953488378
```

Load data from a dictionary, train model, and save predictions to a new DataFrame. Predict car prices based on age (years): Age [1, 2, 3, 4, 5], Price [25000, 22000, 19000, 16000, 13000]. Predict for ages [2.5, 3.5, 6].

```python
import pandas as pd
from sklearn import linear_model

data = {
    'age': [1, 2, 3, 4, 5],
    'price': [25000, 22000, 19000, 16000, 13000]
}
df=pd.DataFrame(data)
reg=linear_model.LinearRegression()
reg.fit(df[['age']],df['price'])
```

```python
new_age=[2.5,3.5,6]
predictions=reg.predict([[a] for a in new_age])
predictions
```

Canada Per Capita Income Challenge: Year [1970, 1980, 1990, 2000, 2010], Income [3000, 9000, 15000, 22000, 30000]. Predict income for 2020 and 2025. Also calculate what year income might reach $50,000.

```python
import pandas as pd
from sklearn import linear_model

df = pd.DataFrame({
    'year': [1970, 1980, 1990, 2000, 2010],
    'income': [3000, 9000, 15000, 22000, 30000]
})
new_df=pd.DataFrame(df)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df['income'])
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2025]])
```

```python
m=reg.coef_
c=reg.intercept_
target=50000
year=(target-c)/m
year
```

```python
#read mode 'rb'  read binary mode
# 'model_pickle'  the file where you earlier stored your model
import pickle
with open('model_pickle','rb') as f:
  mp=pickle.load(f)
```
