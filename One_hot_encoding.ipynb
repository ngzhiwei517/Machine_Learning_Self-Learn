# Exercise

Create dummy variables from the 'town' column using pandas get_dummies().


```python
import pandas as pd

df = pd.read_csv("homeprices.csv")
df
```

```python
dummy=pd.get_dummies(df.town,dtype=int)
dummy
```

Merge the dummy variables with the original dataframe horizontally using pd.concat().

```python
merge=pd.concat([df,dummy],axis='columns')
merge
```

Drop the original 'town' column from the merged dataframe to create the final dataset.

```python
final=merge.drop(columns='town')
final
```

Avoid the dummy variable trap by dropping one of the dummy columns (e.g., 'west windsor').

```python
final=final.drop(columns='west windsor')
final
```
