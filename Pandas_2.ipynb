# 3_different_ways_of_creating_dataframe

## Using tuples list

```python
weather_data = [
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]
a_df=pd.DataFrame(weather_data,columns=['Date','Temperature','Windspeed','Event'])
a_df
```

## Using list of dictionaries

```python
weather_data = [
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},

]
b_df=pd.DataFrame(weather_data)
b_df
```

# Tutorial 4 Read and Write CSV file

```python
import pandas as pd

df = pd.read_csv("stock_data.csv")
df
```

### We want to skip the header

```python
df = pd.read_csv("stock_data.csv",skiprows=1)
df
```

### Our header starts at row number 1

```python
df = pd.read_csv("stock_data.csv",header=1)
df
```

### When there is no header

```python
import pandas as pd

df = pd.read_csv("stock_data.csv")
df
```

### Provide name by using names

```python
import pandas as pd

df = pd.read_csv("stock_data.csv",names=['tickers','eps','revenue','price','people'])
df
```

### Read certain number of rows

```python
import pandas as pd

df = pd.read_csv("stock_data.csv",nrows=3) #read 3 rows
df
```

### Convert n.a to NAN

```python
import pandas as pd

df = pd.read_csv("stock_data.csv",na_values=['n.a.','non available']) #whenever encounting n.a. or non available, it will converts to NAN
df
```

### Convert -1 in revenue to NAN only (can be done using dict)

```python
import pandas as pd

df = pd.read_csv("stock_data.csv",na_values={
    'eps':['not available','n.a.'],
    'revenue':['-1','n.a.'],
    'people':['n.a.']
})
df
```

## Saving data frame as csv

#### if don't want index can set index=False

```python
df.to_csv('new.csv')
```

#### If we only want to save specific column

```python
df.to_csv('new.csv',columns=['eps','revenue'])
```

### Convert string to date datatype using parse_dates

```python
import pandas as pd
df=pd.read_csv("weather_data.csv",parse_dates=['day'])
df.set_index('day',inplace=True)
```

```python
df.loc['2017-01-01']
```

```python
new_df=df.fillna(0)
new_df
```

### Set N.A to specific value

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```
