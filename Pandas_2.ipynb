# 3_different_ways_of_creating_dataframe

## Using tuples list

```python
weather_data = [
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]
a_df=pd.DataFrame(weather_data,columns=['Date','Temperature','Windspeed','Event'])
a_df
```

# Tutorial 5: Handle missing data (fillna, dropna, interpolate)

```python
import pandas as pd
df=pd.read_csv("weather_data.csv")
df
```

```python
type(df.day)
```

```python
type(df.day[0])
```

### Convert string to date datatype using parse_dates

```python
import pandas as pd
df=pd.read_csv("weather_data.csv",parse_dates=['day'])
df.set_index('day',inplace=True)
```

```python
df.loc['2017-01-01']
```

```python
new_df=df.fillna(0)
new_df
```

### Set N.A to specific value

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

### There is some problem there so wehn we find mean value,due to the existing 0, it might affect the accuracy of the results

#### Set it as previous value (estimate)

```python
new_df=df.fillna(method='ffill') #forward fill
new_df                           # can also use bfill (back fill)
```

```python
new_df
```

```python
new_df=df.fillna(method='bfill')
new_df
```

```python
b_df=df.fillna(method='ffill')
b_df
```

### If we want to copy the value only once from the front and not continuously, then we use limit

```python
df
```

```python
c_df=df.fillna(method='ffill',limit=2)
```

```python
c_df
```

 What is interpolate?

Interpolate = estimate a value in between known data points.

Think of it like filling in the gap smoothly between two values.

 Formula-ish idea:

If you know points A and B, interpolation finds a point between A and B.

```python
df
```

```python
df_inter=df.interpolate()
df_inter
```

image.png

We can notice that interpolate gives us the median value but when we look at the date, it is closer to fifth of Jan, so the value should be closer to 5th of Jan

```python
df_inter=df.interpolate(method='time')
df_inter
```

image.png

## We can use dropna also instead of fillna

```python
new_df=df.dropna()
new_df
```

### If we want to drop the row with all missing values only

```python
new_df=df.dropna(how='all')
new_df
```

### If want to specify like:
we want to keep rows with only one NAN or two NAN (use threshold)

```python
new_df=df.dropna(thresh=2) #we need at least 2 values to keep that row (max one NAN)
new_df
```

```python
import pandas as pd
df = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math': [85, None, 92, 78],
    'Science': [90, 88, None, 85],
    'English': [None, 92, 88, None]
})
df
```

Fill missing values with different values for each column: Math  50, Science  60, English  "Absent".

```python
new_df=df.fillna({
    'Math':50,
    'Science':60,
    'English':'Absent'
})
new_df
```

Fill missing temperature values using forward fill (carry the previous value forward).

```python
import pandas as pd
df = pd.DataFrame({
    'Day': range(1, 8),
    'Temperature': [22, None, None, 25, None, 23, 24]
})
df
```

```python
new_df=df.fillna(method='ffill')
new_df
```

Fill missing values using backward fill, but only fill 1 consecutive missing value

```python
import pandas as pd
df = pd.DataFrame({
    'Value': [10, None, None, None, 20, None, 30]
})

df
```

```python
new_df=df.fillna(method='bfill',limit=1)
new_df
```

Use linear interpolation to fill missing temperature values.

```python
import pandas as pd
df = pd.DataFrame({
    'Day': range(1, 8),
    'Temperature': [20, None, None, 26, None, None, 32]
})
new_df=df.interpolate()
new_df
```

Drop all rows that have at least 2 missing values.
