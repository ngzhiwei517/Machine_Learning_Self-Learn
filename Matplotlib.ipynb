# Matplotlib Tutorial 2 - format strings in plot function

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'g+') #this means green and + marker
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'g+--') #this means green and + marker
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'b*-') #this means blue and * marker - means solid line
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'b*:') #this means green and + marker
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'ro-',markersize=20)
```

### alpha property to control transparency of line chart

```python
plt.plot(x,y,'+-',markersize=20,alpha=0.3) # alpha can be specified on a scale 0 to 1
```

# **Matplotlib Tutorial 3 - Axes labels, Legend, Grid**

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

### Axes labels and chart title

Legend

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.legend(loc='lower left')
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.legend(loc='best')
```

### Set font size

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.legend(loc='best',fontsize='large')
```

### Set grid

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.legend(loc='best')
plt.grid()
```

## Matplotlib Tutorial 4 - Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Simple bar chart showing revenues of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue,label='Revenue')
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```


Multiple Bars showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]
```

```python
xpos=np.arange(len(profit))
xpos
```

```python
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='profit')

plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.barh(xpos,revenue,label='Revenue')
plt.yticks(xpos,company)
plt.ylabel('Revenue(Bln)')
plt.title('US Technology Stocks')
plt.legend()
```
