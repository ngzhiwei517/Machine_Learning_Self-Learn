# Matplotlib Tutorial 2 - format strings in plot function

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'g+') #this means green and + marker
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'g+--') #this means green and + marker
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'b*-') #this means blue and * marker - means solid line
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'b*:') #this means green and + marker
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'ro-',markersize=20)
```

### alpha property to control transparency of line chart

```python
plt.plot(x,y,'+-',markersize=20,alpha=0.3) # alpha can be specified on a scale 0 to 1
```
