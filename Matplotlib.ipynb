# **Matplotlib Tutorial 3 - Axes labels, Legend, Grid**

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

### Axes labels and chart title

Legend

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.legend(loc='lower left')
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.legend(loc='best')
```

### Set font size

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.legend(loc='best',fontsize='large')
```

### Set grid

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.legend(loc='best')
plt.grid()
```

## Matplotlib Tutorial 4 - Bar Chart

image.png

ðŸ“Š Bar chart

Data type: Categorical (e.g. major, gender, product type)

Bars: Separated (gaps between bars)

Order: Any order (can be rearranged)

ðŸ“ˆ Histogram

Data type: Numerical (continuous) (e.g. height, marks, time)

Bars: Touch each other (no gaps)

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Simple bar chart showing revenues of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue,label='Revenue')
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```


Multiple Bars showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]
```

```python
xpos=np.arange(len(profit))
xpos
```

```python
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='profit')

plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.barh(xpos,revenue,label='Revenue')
plt.yticks(xpos,company)
plt.ylabel('Revenue(Bln)')
plt.title('US Technology Stocks')
plt.legend()
```

# Exercise

Create a grouped bar chart comparing Q1 and Q2 sales for 4 stores. Store A: Q1=100, Q2=120; Store B: Q1=90, Q2=95; Store C: Q1=110, Q2=130; Store D: Q1=85, Q2=100.

```python
import matplotlib.pyplot as plt
import numpy as np

stores = ['Store A', 'Store B', 'Store C', 'Store D']
q1_sales = [100, 90, 110, 85]
q2_sales = [120, 95, 130, 100]
```

```python
xpos=np.arange(len(q1_sales))
xpos
width = 0.35
```

```python
plt.bar(xpos-width/2,q1_sales,width=0.4,label='q1_sales')
plt.bar(xpos+width/2,q2_sales,width=0.4,label='q2_sales')
plt.xticks(xpos,stores)
plt.title('Quarterly Sales Comparison')
plt.ylabel('Sales')
plt.legend()
```

Create a grouped bar chart with 3 groups (Revenue, Costs, Profit) for 4 quarters. Q1: [100, 70, 30], Q2: [120, 80, 40], Q3: [110, 75, 35], Q4: [140, 85, 55].

```python
import matplotlib.pyplot as plt
import numpy as np

quarters = ['Q1', 'Q2', 'Q3', 'Q4']
revenue = [100, 120, 110, 140]
costs = [70, 80, 75, 85]
profit = [30, 40, 35, 55]
x_pos=np.arange(len(revenue))
x_pos
width=0.25
```

```python
plt.bar(x_pos-width,revenue,width=width,label='revenue')
plt.bar(x_pos,costs,width=width,label='costs')
plt.bar(x_pos+width,profit,width=width,label='profit')
plt.xticks(x_pos,quarters)
plt.ylabel('Amount ($M)')
plt.title("Quarterly Financial Data")
plt.legend()
```

Create a bar chart with custom colors for each bar. Use data: Categories ['A', 'B', 'C', 'D'] with values [25, 40, 30, 55]. Use colors: ['red', 'green', 'blue', 'orange'].

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 55]
colors = ['red', 'green', 'blue', 'orange']
```

```python
x_pos=np.arange(len(values))
plt.bar(x_pos,values,width=0.4,label='values',color=colors)
plt.xticks(x_pos,categories)
plt.title('Colored Bar Chart')
plt.legend()
```

Create a bar chart with error bars showing sales data with uncertainty. Sales: [50, 75, 60, 90] with errors: [5, 8, 6, 10].

Use the yerr parameter in plt.bar() to add error bars.

```python
import matplotlib.pyplot as plt
import numpy as np

products = ['A', 'B', 'C', 'D']
sales = [50, 75, 60, 90]
errors = [5, 8, 6, 10]
x_pos = np.arange(len(products))
plt.bar(x_pos,sales,yerr=errors,capsize=5)
plt.xticks(x_pos,products)
plt.title('Sales with Error Bars')
```

Create a horizontal stacked bar chart showing time spent on activiti

```python
import matplotlib.pyplot as plt
import numpy as np

days = ['Monday', 'Tuesday', 'Wednesday']
work = [8, 7, 9]
sleep = [7, 8, 7]
leisure = [6, 7, 5]
```

```python
y_pos=np.arange(len(work))
plt.barh(y_pos,work,label='work')
plt.barh(y_pos,sleep,left=np.array(work),label='sleep')
plt.barh(y_pos,leisure,left=np.array(work)+np.array(sleep),label='leisure')
plt.yticks(y_pos,days)
plt.xlabel('Hours')
plt.title("Time spent on different activities")
plt.legend()
```

```python
y_pos=np.arange(len(work))
plt.barh(y_pos-0.2,work,height=0.2,label='work')
plt.barh(y_pos,sleep,height=0.2,label='sleep')
plt.barh(y_pos+0.2,leisure,height=0.2,label='leisure')
plt.yticks(y_pos,days)
plt.title("Time spent on different activities")
plt.legend()
```
