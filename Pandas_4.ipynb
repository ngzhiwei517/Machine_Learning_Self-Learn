# **Pandas Tutorial 10. Pivot table**

ðŸ”„ What is pivot in pandas?

Think of pivot as rearranging your table so that:

One column becomes rows ðŸ§±

One column becomes columns ðŸ“Š

One column becomes values inside the table ðŸ“¦

It reshapes your data from long â†’ wide format.

df.pivot(index='row_col', columns='col_col', values='value_col')

* index â†’ what you want to appear as row labels

* columns â†’ what you want to appear as column labels

* values â†’ the cell values

```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather.csv")
df
```

```python
df.pivot(index='date',columns='city')
```

```python
df.pivot(index='date',columns='city',values='humidity')
```

```python
df.pivot(index='city',columns='date')
```

Use pivot_table() to calculate the average temperature for each city on each date.

```python
weather_duplicates = pd.DataFrame({
    'date': ['5/1/2017', '5/1/2017', '5/2/2017', '5/2/2017',
             '5/1/2017', '5/1/2017'],
    'city': ['new york', 'new york', 'new york', 'new york',
             'mumbai', 'mumbai'],
    'temperature': [65, 61, 70, 72, 75, 78],
    'humidity': [56, 54, 60, 62, 80, 83]
})
weather_duplicates
```

```python
weather_duplicates.pivot_table(index='city',columns='date',values='temperature',aggfunc='mean')
```

####  Group daily temperature data by month.
Calculate average monthly temperature. Group dates by month using pd.Grouper().

```python
time_weather = pd.DataFrame({
    'date': ['2017-05-01', '2017-05-15', '2017-05-25',
             '2017-12-01', '2017-12-15', '2017-12-20'],
    'city': ['new york', 'new york', 'new york',
             'new york', 'new york', 'new york'],
    'temperature': [65, 70, 68, 30, 28, 25]
})
```

```python
time_weather['date']=pd.to_datetime(time_weather['date'])
```

```python
time_weather_df=time_weather.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
time_weather_df
```
