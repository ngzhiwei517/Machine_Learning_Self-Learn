```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
digits = load_digits()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target,test_size=0.3)
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

### Logistic Regression

```python
lr.score(X_test,y_test)
```

### SVM

```python
svm=SVC()
svm.fit(X_train,y_train)
svm.score(X_test, y_test)
```

### Random Forest

```python
rf = RandomForestClassifier(n_estimators=40)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

We notice some problem the number changes everytime we run the model due to different training set and test set

## KFold cross validation

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=4) #  This means 4-fold cross-validation.
```

```python
for train_index,test_index in kf.split([10,9,8,7,6,5,4,3]):
  print(train_index,test_index)
```

üìå Important:
These numbers are positions (indices) of [1,2,3,4,5,6,7,8]

Takes 8 samples

Splits them into 4 folds

Each fold has 2 samples as test data

Returns indices, not values ‚ùó

## Use KFold for our digits example

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
get_score(RandomForestClassifier(),X_train,X_test,y_train,y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
folds=StratifiedKFold(n_splits=3)
score_log=[]
score_svm=[]
score_rf=[]

for train_index,test_index in folds.split(digits.data,digits.target):
  X_train,X_test,y_train,y_test=digits.data[train_index],digits.data[test_index],digits.target[train_index],digits.target[test_index]
  score_log.append(get_score(LogisticRegression(),X_train,X_test,y_train,y_test))
  score_svm.append(get_score(SVC(gamma='auto'), X_train, X_test, y_train, y_test))
  score_rf.append(get_score(RandomForestClassifier(n_estimators=40),X_train,X_test,y_train,y_test))
```

```python
score_log
```

```python
score_svm
```

```python
score_rf
```

in real life can just use cross_val score

```python
from sklearn.model_selection import  cross_val_score
cross_val_score(LogisticRegression(),digits.data,digits.target,cv=3) #by default is 5
```

```python
from sklearn.model_selection import  cross_val_score
cross_val_score(SVC(gamma='auto'),digits.data,digits.target,cv=3) #by default is 5
```

```python
from sklearn.model_selection import  cross_val_score
cross_val_score(RandomForestClassifier(),digits.data,digits.target,cv=3) #by default is 5
```

Load the wine dataset and perform a simple train-test split with one model to establish a baseline.

Use cross_val_score to evaluate the same Random Forest on wine dataset with 5-fold CV. Compare with single split.

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

wine = load_wine()
X, y = wine.data, wine.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

When using cross_val_score, X should be the full dataset (all your data), not just X_train or X_test.

cross_val_score performs its own splitting internally:

By default, it does 5-fold cross-validation on the entire X and y you pass

It automatically creates train/validation folds

It returns scores on the validation folds (not your separate test set

```python
cross_val_score(RandomForestClassifier(n_estimators=100,random_state=42),X,y)
```
