```python

import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradient_descent(x,y):
  m_curr=b_curr=0
  iterations=10000
  n=len(x)
  learning_rate=0.001 #trial and error see how the algo behaves and fine tune it
  for i in range(iterations):
    y_predicted=m_curr*x+b_curr
    cost=1/n*sum(x**2 for x in (y-y_predicted))
    md=-(2/n)*sum(x*(y-y_predicted))
    bd=-(2/n)*sum(y-y_predicted)
    m_curr=m_curr-learning_rate*(md)
    b_curr=b_curr-learning_rate*(bd)
    print(f'm:{m_curr},b:{b_curr},cost:{cost},iterations:{i}')
```

```python
gradient_descent(x,y)
```

```python
import math
def gradient_descent(x,y):
  x = np.array(x)
  y = np.array(y)
  m_curr=b_curr=0
  iterations=100000
  learning_rate=0.002
  cost_previous=0
  n=len(x)
  for i in range(iterations):
    y_predicted=m_curr*x+b_curr
    cost=1/n*sum(k**2 for k in (y-y_predicted))
    md=-(2/n)*sum(x*(y-y_predicted))
    bd=-(2/n)*sum(y-y_predicted)
    if math.isclose(cost,cost_previous,rel_tol=1e-20):
      break
    cost_previous=cost
    m_curr=m_curr-learning_rate*md
    b_curr=b_curr-learning_rate*bd
    print(f'm:{m_curr},b: {b_curr},cost:{cost},iterations:{i}')
  return m_curr, b_curr
```

```python
m,b=gradient_descent(x,y)
```

```python
print(f"Using gradient descent function:m:{m},b:{b}")
```

 1. What are Dummy Variables?

Dummy variables = Encoding categorical data into 0/1 columns, BUT you drop ONE column to avoid multicollinearity.

```python
dummies=pd.get_dummies(df.town,dtype=int)
dummies
```

```python
#concat can use to join 2 data frame
merge=pd.concat([df,dummies],axis='columns')
merge
```

```python
final=merge.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X=final.drop('price',axis='columns')
X
```

```python
y=final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
#model.score(X_test, y_test) returns the accuracy
model.score(X,y)
```

# Using sklearn OneHotEncoder

 2. What is One-Hot Encoding?

One-hot encoding = encoding categorical data into 0/1 columns WITHOUT dropping anything.
 It keeps all categories.

 When to Use One-Hot Encoding (k columns)

Use one-hot encoding for most machine learning models, especially those not harmed by multicollinearity.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
new_df=df
#take label column as input and return label column
new_df.town=le.fit_transform(new_df.town)
```
