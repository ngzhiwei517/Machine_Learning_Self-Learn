Handling Missing Values

```python
# Creating a DataFrame with missing values from a list of lists
data = [['Geek1', 28, 'Engineer'],
        ['Geek2', None, 'Data Scientist'],
        ['Geek3', 32, None]]
f_df=pd.DataFrame(data,columns=['Name','Age','Occupation'])
f_df=f_df.replace({None:'NaN'})
f_df
```

DataFrame With Different Data Types

```python
data = [['Geek1', 28, 'Engineer'],
        ['Geek2', 25, 'Data Scientist'],
        ['Geek3', '32', 'Manager']]  # Age represented as a string
f_df=pd.DataFrame(data,columns=['Name','Age','Occupation'])
f_df['Age']=pd.to_numeric(f_df['Age'])
f_df
```

#  What is a Pandas Series?

A Series = one column of data

It is like a single column of a DataFrame, but by itself.

 A Series has 2 parts:

* Values (the data)

* Index (the labels for each value)

```python
s=pd.Series([10,20,30,50])
s
```

 Why do we need a Series?

Because a DataFrame is made of many Series.

image.png

```python
a=pd.Series([1,2,3],index=['a','b','c'])
a
```

```python
pd.Series({'d':1,'e':2,'f':3})
```

```python
# Creating a list
author = ['Jitender', 'Purnima',
          'Arpit', 'Jyoti']
auth_series=pd.Series(author)
auth_series
```

```python
# Creating two lists
author = ['Jitender', 'Purnima',
          'Arpit', 'Jyoti']
article = [210, 211, 114, 178]
auth_series=pd.Series(author)
article_series=pd.Series(article)
frame={'Author':auth_series,'Article':article_series}
result=pd.DataFrame(frame)
result
```

```python
age = [21, 21, 24, 23]
result['Age']=pd.Series(age)
result
```
