#### We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
import matplotlib.pyplot as plt
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar)
```

[77, 84.3, 91.6, 98.9, 106.2, ..., 150]
These numbers are the edges of the bins.

So the bins are:

Bin 1: 77  84.3

Bin 2: 84.3  91.6

Bin 3: 91.6  98.9

Bin 4: 98.9  106.2

...

Last bin: 142.7  150

Each bar in the histogram represents one of these ranges.



---



This means:

3 values fall between 7784.3

3 values fall between 84.391.6

1 value falls between 91.698.9

0 values between 98.9106.2

```python
plt.hist(blood_sugar,bins=3)
```

rwidth = 1.0  bars touch each other (full bin width)

rwidth = 0.8  bars use 80% of the bin width (small gaps)

rwidth = 0.5  thinner bars, bigger gaps

```python
plt.hist(blood_sugar,bins=3,rwidth=0.8)
```

## Customize the bin range

```python
plt.xlabel('Blood sugar level')
plt.ylabel('Number of patients')
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.8,color='g')
```

```python
plt.ylabel('Blood sugar level')
plt.xlabel('Number of patients')
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.8,color='g',orientation='horizontal',label='Num of patients')
plt.legend()
```

# Matplotlib Tutorial 6 - Pie Chart

image.png

```python
import matplotlib.pyplot as plt
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.axis('equal')
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=0.8)
```

### Show percentages for every pie. Specify radius to increase chart size

This is a format string.

%  start of formatting

1  minimum width (not very important here)

.1  1 decimal place

f  floating-point number

%%  print a % sign

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct='%1.1f%%')
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct='%1.2f%%')
```

## Explode

It separates (pulls out) slices from the center to highlight them.

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct='%1.2f%%',explode=[0,0.1,0.1,0,0])
```

# Exercise

Create a simple pie chart with expenses: [1400, 600, 300, 410, 250] and labels: ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"].

* Add percentages to each slice using autopct='%1.1f%%'.

* "Explode" the last two slices using explode=[0, 0, 0, 0.1, 0.2] to separate them from the center

* Change direction to counterclockwise and rotate starting angle to 45 degrees using counterclock=True, startangle=45.

```python
import matplotlib.pyplot as plt

exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,autopct='%1.1f%%',explode=[0,0,0,0.1,0.2],counterclock=True,startangle=45)
plt.axis('equal')
plt.savefig('piechart.png')
```

# Matplotlib Tutorial 7 - Save Chart To a File Using savefig

```python
import matplotlib.pyplot as plt

exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,autopct='%1.1f%%',explode=[0,0,0,0.1,0.2],counterclock=True,startangle=45)
plt.axis('equal')
plt.savefig('piechart.png',bbox_inches='tight',pad_inches=2) #bbox_inches:fit the size of the diagram very well
# pad_inches: extra inches of the image
```

### Can save the piechart as pdf also

```python
import matplotlib.pyplot as plt

exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,autopct='%1.1f%%',explode=[0,0,0,0.1,0.2],counterclock=True,startangle=45)
plt.axis('equal')
plt.savefig('piechart.pdf',bbox_inches='tight',pad_inches=2,transparent=True) # making the background transparent
```
